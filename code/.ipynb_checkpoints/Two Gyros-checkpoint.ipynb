```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class Vector3:
    
    def __init__(self, size):
        self.x = np.zeros(size)
        self.d = np.zeros(size)
        self.d2 = np.zeros(size)
        self.size = size
        
    def __getitem__(self, key):
        return self.x[key]
    
    def __setitem__(self, key, value):
        self.x[key] = value
    
def integrate_single(f_a, x, t):
    dt = t[1]-t[0]
    for i in range(x.size-1):
        x[i+1] = x[i] + x.d[i]*dt + 0.5*f_a(x[i], x.d[i], t[i])*dt**2
        x.d[i+1] = x.d[i] + 0.5*(f_a(x[i], x.d[i], t[i]) + f_a(x[i+1], x.d[i+1], t[i+1]))*dt
    
def integrate_coupled(functions, variables, t):
    dt = t[1]-t[0]
    for i in range(t.size - 1):
        for name in functions.keys():
            x = variables[name]
            f = functions[name]
            x[i+1] = x[i] + x.d[i]*dt + 0.5*f(variables, i, t[i])*dt**2
        for name in functions.keys():
            x = variables[name]
            f = functions[name]
            x.d[i+1] = x.d[i] + 0.5*(f(variables, i, t[i]) + f(variables, i+1, t[i+1]))*dt

            
```

```python
n = 100
x = Vector3(n)
t = np.linspace(0, 10, n)

omega = 2
damping = 0.2
target = 1
f_a = lambda x, xd, t: omega**2*(target-x) - 2*damping*omega*xd

# Analytical:
omega_d = omega*np.sqrt(1 - damping**2)
psi = np.arcsin(damping)
y = target*(1 - np.exp(-damping*omega*t)*np.cos(omega_d*t - psi)/np.cos(psi))

integrate_single(f_a, x, t)

plt.plot(t, x.x)
plt.plot(t, y)
```

```python
psi_p = 10000 * (2*np.pi/60)
psi_q = 5000 * (2*np.pi/60)

m = 5
g = 9.81
d = 10 * 10e-2
C = 0.1

k = np.hypot(psi_p, psi_q)
epsilon = np.arctan(psi_q/psi_p)
print(np.degrees(epsilon)-45)

alpha = np.linspace(0, 0.75*np.pi, 100)

num = np.sqrt(2)*m*g*d*np.sin(alpha)
den = C*k*np.sin(alpha - epsilon + np.pi/4)
phidot = num/den

plt.plot(np.degrees(alpha), phidot)
plt.xlabel(r'$\alpha$ (degrees)')
plt.ylabel(r'$\dot{\phi}$ (rad/s)')
plt.show()
```

$$ \phi = \frac{mgd}{C\dot{\psi}_{\textrm{mean}}}f(\alpha) $$

$$ \textrm{Precession Factor } f(\alpha) = \frac{\sqrt{2}\sin{(\alpha)}}{\sin{(\alpha - \alpha_{\textrm{crit}})}} $$

$$ \textrm{Geometric Mean Spin  } \dot{\psi}_{\textrm{mean}} = \sqrt{\dot{\psi}_1 + \dot{\psi}_2} $$

$$ \textrm{Critical Angle  } \alpha_{\textrm{crit}} = \arctan{\left(\frac{1}{k}\right)} - \frac{\pi}{4} $$

$$ \textrm{Spin Ratio  } k = \dot{\psi}_1 / \dot{\psi}_2 $$

If $\dot{\psi}_1 = \dot{\psi}_2$, then precession rate doesn't depend on the inclination angle $\alpha$, but gives constant precession, equivalent to a single gyroscope with $\dot{\psi}^2 = \dot{\psi}_1^2 + \dot{\psi}_2^2$

Otherwise, the precession factor gives a variation in precession rate with inclination. <br>
At $\alpha = \alpha_{\textrm{crit}} + n\pi$, $\phi \to \infty$. <br>
This defines a region where precession rate is bounded:
- For $k>1$, $\alpha_{\textrm{crit}}<0$, so is bounded for: $-|\alpha_{\textrm{crit}}| < \alpha < \pi - |\alpha_{\textrm{crit}}|$
- For $k<1$, $\alpha_{\textrm{crit}}0$, so is bounded for: $-\pi + \alpha_{\textrm{crit}} < \alpha < \alpha_{\textrm{crit}}$

In other words, if $\alpha$ is measured in the direction of the gyroscope with the larger $\dot{\psi}$ and we take $\alpha_{\textrm{crit}}$ as $|\alpha_{\textrm{crit}}|$, then $\phi$ is bounded for: $-\alpha_{\textrm{crit}} < \alpha < \pi - \alpha_{\textrm{crit}}$ 

```python
def alpha_crit(k):
    return np.arctan(1/k) - np.pi/4

def precession_factor(alpha, k):
    alpha_crit = np.arctan(1/k) - np.pi/4
    return np.sqrt(2)*np.sin(alpha) / np.sin(alpha - alpha_crit)
```

```python
ks = np.linspace(1, 10, 100)
alpha_crit = np.abs(alpha_crit(ks))
lower_limit = np.degrees(-alpha_crit)
upper_limit = np.degrees(np.pi - alpha_crit)
plt.plot(ks, lower_limit, c='r')
plt.plot(ks, upper_limit, c='r')
plt.fill_between(ks, lower_limit, upper_limit, color='#EEEEEE')
plt.show()
```

```python
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator

#x = alpha
#y = k
min_alpha = 0.0001
max_alpha = 0.74*np.pi
min_k = 1
max_k = 3

# make these smaller to increase the resolution
dx, dy = 0.05, 0.05

# generate 2 2d grids for the x & y bounds
y, x = np.mgrid[slice(min_k, max_k + dy, dy),
                slice(min_alpha, max_alpha + dx, dx)]

#z = np.array([[precession_factor(x[i, j], y[i, j]) for j in range(x.shape[1])]
#     for i in range(x.shape[0])])
alpha_crit = np.arctan(1/y) - np.pi/4
z = np.sqrt(2)*np.sin(x) / np.sin(x - alpha_crit)

# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
levels = MaxNLocator(nbins=50).tick_values(z.min(), z.max())


# pick the desired colormap, sensible levels, and define a normalization
# instance which takes data values and translates those into levels.
cmap = plt.get_cmap('viridis')
norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

fig, ax = plt.subplots(1)

# contours are *point* based plots, so convert our bound into point
# centers
cf = ax.contourf(np.degrees(x[:-1, :-1] + dx/2.),
                  y[:-1, :-1] + dy/2., z, levels=levels,
                  cmap=cmap)
fig.colorbar(cf, ax=ax)
ax.set_title('precession factor')
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'k')
# adjust spacing between subplots so `ax1` title and `ax0` tick labels
# don't overlap
fig.tight_layout()

plt.show()
```
