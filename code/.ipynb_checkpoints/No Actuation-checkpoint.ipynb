```python
import numpy as np
import matplotlib.pyplot as plt

class Properties:
    
    def __init__(self, A=0, B=0, C=0):
        self.A = A
        self.B = B
        self.C = C
        
    def unpack(self):
        return self.A, self.B, self.C
    
def simulate(properties, initial_omega, T, N, Q_func = None):
    
    A, B, C = properties.unpack()

    def f(t, omega):
        if Q_func is None:
            Q = np.zeros(3)
        else:
            Q = Q_func(t)
        omega_dot_1 = Q[0]/A - omega[1]*omega[2] * (C-B)/A
        omega_dot_2 = Q[1]/B - omega[0]*omega[2] * (A-C)/B
        omega_dot_3 = Q[2]/C - omega[0]*omega[1] * (B-A)/C
        return np.array([omega_dot_1, omega_dot_2, omega_dot_3])
    
    omega = np.zeros((N, 3))
    omega[0] = initial_omega

    t = np.linspace(0, T, N)
    dt = T/(N-1)

    for n in range(N-1):
        k1 = f(t[n], omega[n])
        k2 = f(t[n] + dt/2, omega[n] + k1*dt/2)
        k3 = f(t[n] + dt/2, omega[n] + k2*dt/2)
        k4 = f(t[n] + dt, omega[n] + k3*dt)
        omega[n+1] = omega[n] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)*dt
    
    if Q_func is None:
        return t, omega
    else:
        return t, omega, np.array([Q_func(ti) for ti in t])
    
def plot_simulation(t, omega, name=None):
    
    plt.figure(figsize=(9, 5))

    plt.plot(t, omega[:, 0], label=r'$\Omega_1$')
    plt.plot(t, omega[:, 1], label=r'$\Omega_2$')
    plt.plot(t, omega[:, 2], label=r'$\Omega_3$')
    plt.legend(loc = 'lower right')
    plt.xlabel('Time t (s)', fontsize=12)
    plt.ylabel(r'$\mathbf{\Omega}$ (rad/s)', fontsize=12, labelpad=-4)

    if name is not None:
        plt.savefig('../fig/{}'.format(name), bbox_inches='tight')
    plt.show()
    
def plot_simulation_with_couple(t, omega, couple, name=None):
    
    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(6, 5))
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.25)

    ax0.plot(t, couple[:, 0], label=r'$Q_1$')
    ax0.plot(t, couple[:, 1], label=r'$Q_2$')
    ax0.plot(t, couple[:, 2], label=r'$Q_3$')

    ax0.legend(loc = 'lower right')
    ax0.set_xlabel('Time t (s)', fontsize=12)
    ax0.set_ylabel(r'$\mathbf{Q}$ (Nm)', fontsize=12)

    ax1.plot(t, omega[:, 0], label=r'$\Omega_1$')
    ax1.plot(t, omega[:, 1], label=r'$\Omega_2$')
    ax1.plot(t, omega[:, 2], label=r'$\Omega_3$')
    ax1.legend(loc = 'lower right')
    ax1.set_xlabel('Time t (s)', fontsize=12)
    ax1.set_ylabel(r'$\mathbf{\Omega}$ (rad/s)', fontsize=12, labelpad=-4)
    
    if name is not None:
            plt.savefig('../fig/{}'.format(name), bbox_inches='tight')
    plt.show()
```

# ABC, constant rotation, stable

```python
properties = Properties(100, 150, 200)
initial_omega = np.array([10, 0.1, -0.2])
T = 10
N = 100

t, omega = simulate(properties, initial_omega, T, N)

plot_simulation(t, omega, 'no_actuation_ABC_stable.png')
```

# Constant rotation, no spin, AAA body

```python
properties = Properties(100, 150, 200)
initial_omega = np.array([0.1, 10, -0.2])
T = 10
N = 100

t, omega = simulate(properties, initial_omega, T, N)

plot_simulation(t, omega, 'no_actuation_ABC_unstable.png')
```

# Stable motion about a single axis with variable angular velocity

```python
properties = Properties(100, 150, 200)
initial_omega = np.array([0, 0.01, -0.005])
T = 10
N = 100

def Q_func(t):
    if t<1 or t>6:
        return np.zeros(3)
    else:
        Q = 20*np.sin(2*np.pi*(t-1)/5)
        return np.array([Q, 0, 0])

t, omega, Q = simulate(properties, initial_omega, T, N, Q_func)

plot_simulation_with_couple(t, omega, Q, 'no_actuation_ABC_couple.png')
```
